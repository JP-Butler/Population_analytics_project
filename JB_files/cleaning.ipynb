```python
# Import pandas for data cleaning 

import pandas as pd
```

```python
first_df =  pd.read_csv("life_expect.csv", header=3)
first_df
```

```python
# 2019-2021 life expectancy df
columns = ["Country Name","Country Code","Indicator Name","2019", "2020", "2021"]
life_expect_df = first_df[columns]
life_expect_df
```

```python
# 2021 year only 
columns_2021 = ["Country Name","Country Code","Indicator Name", "2021"]

life_expect_df2021 = life_expect_df[columns_2021]
life_expect_df2021
```

```python
# Read in file with 2021 Mortality Rates

mort_df = pd.read_csv('Mortality_Rates.csv', header=3)
mort_df
```

```python
mort_columns = ["Country Name","Country Code","Indicator Name", "2021"]
revised_mort_df = mort_df[mort_columns]
revised_mort_df
```

```python
life_expect_df2021
```

```python
life_mort_merge_df = pd.merge(life_expect_df2021, revised_mort_df, how='outer', left_on='Country Name', right_on='Country Name', suffixes=['_life_expect', '_mortality'])
life_mort_merge_df.loc[life_mort_merge_df[['Country Code_life_expect', 'Country Code_mortality']].isnull().any(axis=1)]
```

```python
translate_life = {"Yemen, Rep.": "Yemen","Brunei Darussalam":"Brunei", "Cote d'Ivoire":"Cote D'Ivoire", "Congo, Dem. Rep.":"DR Congo", "Congo, Rep.":"Congo", "Czechia":"Czech Republic (Czechia)",
                  "Micronesia, Fed. Sts.":"Micronesia","Kyrgyz Republic":"Kyrgyzstan", "Korea, Rep.":"South Korea", "Lao PDR":"Laos","Korea, Dem. People's Rep.":"North Korea",
                    "Russian Federation":"Russia", "Sao Tome and Principe":"Sao Tome & Principe","Sint Maarten (Dutch part)": "Sint Maarten", "Turks and Caicos Islands":"Turks and Caicos", 
                     "St. Vincent and the Grenadines":"St. Vincent & Grenadines", "Venezuela, RB": "Venezuela", "Virgin Islands (U.S.)":"U.S. Virgin Islands"}
```

```python
life_expect_df2021['Country Name'] = life_expect_df2021['Country Name'].replace(translate_life)
revised_mort_df['Country Name'] = revised_mort_df['Country Name'].replace(translate_life)
```

```python
life_mort_merge_df = pd.merge(life_expect_df2021, revised_mort_df, how='outer', left_on='Country Name', right_on='Country Name', suffixes=['_life_expect', '_mortality'])
life_mort_merge_df.loc[life_mort_merge_df[['Country Code_life_expect', 'Country Code_mortality']].isnull().any(axis=1)]
```

```python
# dropped uneeded columns 
life_mort_merge_df
```

```python
# rename Country Name colunn to match 'clean_wp2.csv'
new_columns = {"Country Name":"Country (or dependency)",
               "2021_life_expect":"Life Expectency (2021)" ,
               "2021_mortality":"Mortality Rate (2021)"}

final_life_mort_df = life_mort_merge_df.rename(columns=new_columns)
final_life_mort_df
```

```python
clean_wp_df = pd.read_csv('clean_wp2.csv')
clean_wp_df
```

```python
fully_merged_df = pd.merge(clean_wp_df, final_life_mort_df, on='Country (or dependency)',how='left')
fully_merged_df
```

```python
# Save final merged df to a csv

fully_merged_df.to_csv('merged_clean_wp.csv')
```

```python
# Check null counts for 2 new columns

mort_nulls = fully_merged_df['Mortality Rate (2021)'].isnull().sum()
life_nulls = fully_merged_df['Life Expectency (2021)'].isnull().sum()

print("Mortality Rate null count:", mort_nulls)
print("Life Expectency null count:", life_nulls)
```
