```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

# File to Load 
world_population = Path("All Countries.xlsx")

world_population_data = pd.read_excel(world_population)

world_population_data
```

```python
# Seperate needed data
clean_wp = world_population_data[['Country (or dependency)', 'Population  (2023)', 'Yearly  Change',
       'Net  Change', 'Land Area  (Km²)', 'Fert.  Rate', 'Med.  Age', 'World  Share']]

clean_wp
```

```python
# Change the columns "Yearly Change" & "World Share" to a percentage
clean_wp['Yearly  Change'] = clean_wp['Yearly  Change'] * 100

clean_wp['World  Share'] = clean_wp['World  Share'] * 100

# Rename multiple columns
clean_wp = clean_wp.rename(columns={'Yearly  Change': 'Yearly  Change %', 'Fert.  Rate': 'Fertility Rate', 'Med.  Age': 'Median Age', 'World  Share': 'World  Share %'})

# Display the updated DataFrame
clean_wp
```

```python
# Export the clean_wp into a csv
clean_wp.to_csv("clean_data/clean_wp.csv", index = False)
```
