```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
import os
import pandas as pd

# Get the current directory
current_directory = os.getcwd()

# Get the list of all files in the current directory
file_list = os.listdir(current_directory)

# Filter only files with the ".xlsx" extension
excel_files = [file for file in file_list if file.endswith('.xlsx')]
avidThesesFiles=['All Countries.xlsx','World Population.xlsx']

# Print the list of Excel files
print(excel_files)

# File to Load 
for excel_file in excel_files:
    if excel_file not in avidThesesFiles:
        print(excel_file)
        world_population = Path(excel_file)
        world_population_data = pd.read_excel(world_population)
        print(world_population_data.head())
            # Seperate needed data
        clean_top10 = world_population_data[['Year', 'Population', 'Yearly %   Change', 'Yearly  Change', 'Median Age', 'Fertility Rate', "Country's Share of  World Pop", 'World Population']]

        # Change the columns "Yearly Change" & "World Share" to a percentage
        clean_top10['Yearly %   Change'] = clean_top10['Yearly %   Change'] * 100

        clean_top10["Country's Share of  World Pop"] = clean_top10["Country's Share of  World Pop"] * 100

        # Display the updated DataFrame
        clean_top10
        # Export the clean_wp into a csv
        filePath="clean_data/"+excel_file
        # clean_wp.to_csv("clean_data/clean_wp.csv", index = False)
        clean_top10.to_excel(filePath, index=False)
```
