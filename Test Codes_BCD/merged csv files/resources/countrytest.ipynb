```python
import pandas as pd
import csv

# read csv file
country_data_path = "resources/merged_clean_wp.csv"
lat_lon_path = "resources/lat-lon.csv"

country_data = pd.read_csv(country_data_path, encoding="latin1")
lat_lon_data = pd.read_csv(lat_lon_path, encoding="latin1")

# starter dataframe
country_data
```

```python
lat_lon_data
```

```python
lat_lon_data.rename(columns={'country': 'Country (or dependency)', 'latitude': 'Latitude', 'longitude': 'Longitude', 'country_code': 'Country Code'}, inplace=True)
lat_lon_data
```

```python
latlon_columns = ["Country (or dependency)","Latitude","Longitude","Country Code"]
latlon_df = lat_lon_data[latlon_columns]
latlon_df
```

```python
complete_merged = pd.merge(latlon_df, country_data, how="left", on=["Country (or dependency)"])
complete_merged
```

```python
wp_alldata = complete_merged.drop('Unnamed: 0', axis=1)
wp_alldata
```

```python
wp_alldata.rename(columns={'Country (or dependency)': 'Country'}, inplace=True)
wp_alldata
                  
```

```python
wp_alldata.to_csv('resources/allworlddata.csv', index=False)
```
